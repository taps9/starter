```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame con los datos
data = pd.DataFrame({
    'Días': ['Dom-Lun', 'Lun-Mar', 'Mar-Mie', 'Mie-Jue', 'Jue-Vie', 'Vie-Sab', 'Sab-Dom', 'Dom-Lun'],
    'Horas de sueño': [8, 6, 7, 6, 7, 6, 5, 8],
    'Puse alarma en la mañana': ['No', 'No', 'No', 'Si', 'Si', 'Si', 'Si', 'No']
})

# Configurar colores para  alarma
colores_alarma = {'Si': 'lightgreen', 'No': 'lightcoral'}

# Crear un gráfico de barras apiladas
plt.figure(figsize=(10, 6))

plt.bar(data['Días'], data['Horas de sueño'], color='skyblue', label='Horas de sueño')

bottom = [0] * len(data)

bottom = [0] * len(data)

for alarma in set(data['Puse alarma en la mañana']):
    plt.bar(data['Días'], data['Puse alarma en la mañana'].apply(lambda x: 1 if x == alarma else 0), color=colores_alarma[alarma], bottom=bottom, label=f'Alarma: {alarma}')
    bottom = [bottom[i] + data['Puse alarma en la mañana'].apply(lambda x: 1 if x == alarma else 0)[i] for i in range(len(data))]

# Personalizar el gráfico
plt.xlabel('Días de la semana')
plt.ylabel('Horas de sueño /  Alarma por la mañana')
plt.title('Hábitos semanales de sueño y alarma')
plt.xticks(rotation=45)
plt.legend(loc='upper right', bbox_to_anchor=(1.35, 1.0))

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# DataFrame
data = pd.DataFrame({
    'Días': ['Dom-Lun', 'Lun-Mar', 'Mar-Mie', 'Mie-Jue', 'Jue-Vie', 'Vie-Sab', 'Sab-Dom', 'Dom-Lun'],
    'Actividad nocturna': ['Nada', 'Partido de fútbol', 'Ensayo de teatro', 'Ensayo de teatro', 'Ensayo de teatro', 'Nada', 'Función de teatro', 'Función de teatro']
})

# Gráfico de barras interactivo con Plotly
fig = px.bar(data, x='Días', y='Actividad nocturna')

fig.update_layout(
    title='Actividades nocturnas semanales',
    xaxis_title='Días de la semana',
    yaxis_title='Actividad nocturna',
    xaxis={'categoryorder': 'array', 'categoryarray': ['Dom-Lun', 'Lun-Mar', 'Mar-Mie', 'Mie-Jue', 'Jue-Vie', 'Vie-Sab', 'Sab-Dom']},
)
# Mostrar el gráfico interactivo
fig.show()
```
